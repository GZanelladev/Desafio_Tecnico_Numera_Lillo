## Introdução
A gelateria Lillo é muito famosa na capital mineira e vende gelatos de altíssima qualidade com foco no público de classe B+. Quando um cliente compra algum produto, é incentivado que ele preencha seu cpf, email e telefone para receber promoções e novidades. Com esses dados, a gelateria possui uma base de clientes robusta.

A gerente da gelateria pediu a nossa ajuda para melhorar as campanhas de marketing e aumentar a receita. Para isso, ela disponibilizou 3 bases com as seguintes informações:

### Dicionário de Variáveis

**Portfolio_ofertas:** contém as ofertas que são enviadas e possui as seguintes colunas:

· ID: código de identificação da oferta

· Oferta: Tipo de oferta enviada, podendo ser informativo, desconto e compre 1 e leve 2.

· Canal: canal de envio da comunicação podendo ser email, web, celular e redes sociais.

· Valor mínimo: valor mínimo necessário a ser gasto para aproveitar a oferta

· Recompensa: pontuação obtida por completar a compra da oferta

· Duração: tempo de duração da oferta em dias.

**Dados_clientes:** dados demográficos dos clientes e possui as seguintes colunas:

· Gênero: gênero do cliente (F: Feminino, M: Masculino, O: Outro)

· Idade: idade dos clientes

· Id: código único de cada cliente

· Membro_desde: data em que o cliente criou uma conta

· Renda_anual: renda anual fornecida pelo cliente.

**Eventos_ofertas:** base com registros de todos os eventos

· Tipo_evento: descrição dos ofentos ocorridos, podendo ser: transação, oferta recebida, oferta vista, oferta finalizada.

· Cliente: código do cliente que executou aquele evento

· Tempo_decorrido: tempo em horas desde o envio da oferta

· Valor: contém informação do valor gasto na compra

· id_oferta: código da oferta

#### Abordagem para o tratamento de outliers na coluna idade

```python
#Visualização do gráfico boxplot e distribuição das idades dos clientes.

# Criando subplots lado a lado
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# Boxplot de Idade
sns.boxplot(ax=axes[0], y='idade', data=df_clientes)
axes[0].set_title('Boxplot de Idade')

# Histograma (Gráfico de Barras) de Idade
ax = sns.histplot(ax=axes[1], x='idade', data=df_clientes, bins=20, kde=True)
ax.set_title('Histograma de Idade')

# Adicionar linhas para valores médios e mediana
media = df_clientes['idade'].mean()
mediana = df_clientes['idade'].median()

ax.axvline(media, color='red', linestyle='dashed', linewidth=1, label='Média')
ax.axvline(mediana, color='yellow', linestyle='dashed', linewidth=1, label='Mediana')

# Adicionando legenda
ax.legend()

# Ajustando layout para evitar sobreposição
plt.tight_layout()

# Mostrando os gráficos
plt.show()
```

```python
# Calculando os valores do boxplot para a coluna 'idade'
quartis = df_clientes['idade'].quantile([0.25, 0.75])
iqr = quartis[0.75] - quartis[0.25]
limite_superior = quartis[0.75] + 1.5 * iqr

# Imprimindo os valores
print(f"Valor mínimo: {df_clientes['idade'].min()}")
print(f"Primeiro Quartil (Q1): {quartis[0.25]}")
print(f"Mediana (Q2): {df_clientes['idade'].median()}")
print(f"Terceiro Quartil (Q3): {quartis[0.75]}")
print(f"Valor máximo: {df_clientes['idade'].max()}")
print(f"Intervalo Interquartil (IQR): {iqr}")
print(f"Limite Superior (1.5 * IQR): {limite_superior}")

# Imprimindo potenciais outliers
outliers = df_clientes[(df_clientes['idade'] < quartis[0.25] - 1.5 * iqr) | 
                       (df_clientes['idade'] > quartis[0.75] + 1.5 * iqr)]
if not outliers.empty:
    print("Potenciais outliers:")
    print(outliers[['idade']])
else:
    print("Não há potenciais outliers.")
```

```python
# Para verificar a quantidade de outliers 
df_clientes.idade[df_clientes.idade>=115].value_counts()
```

#### Definição da transformação dos dados de idade dos Clientes

- Podemos verificar que existem registros de clientes acima de 100 anos, porém devido a analise por boxplot podemos verificar que o limite superior é de 115 anos, adotando assim 118 anos como outliers.
- Existem 2175 registros de clientes com idade de 118 anos, o que corresponde a 12,8% dos clientes na base de dados.
- É possível que os clientes que não possuem dados de gênero e renda anual sejam exatamente aqueles que possuam a idade mencionada (118 anos). Caso essa suposição seja válida, a abordagem inicial de remover os outliers da base poderia ser considerada, entretanto, é essencial notar que esses clientes também fornecem informações cruciais para a análise, como a data de adesão ao programa de fidelidade e outras possíveis insights que possam surgir.
- Para preservar as características significativas desses clientes, optaremos por substituir o valor de 118 anos pela mediana da distribuição. Ao analisar a distribuição das idades, percebemos que utilizar a média resultaria em um deslocamento da distribuição para a direita.

```python
# Calculando a mediana da coluna 'idade'
mediana_idade = df_clientes['idade'].median()

# Substituindo valores de 118 anos pela mediana
df_clientes.loc[df_clientes['idade'] == 118, 'idade'] = mediana_idade
```

```python
#Visualização do gráfico boxplot e distribuição das idades dos clientes após a transformação.

# Criando subplots lado a lado
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# Boxplot de Idade
sns.boxplot(ax=axes[0], y='idade', data=df_clientes)
axes[0].set_title('Boxplot de Idade')

# Histograma (Gráfico de Barras) de Idade
ax = sns.histplot(ax=axes[1], x='idade', data=df_clientes, bins=20, kde=True)
ax.set_title('Histograma de Idade')

# Adicionar linhas para valores médios e mediana
media = df_clientes['idade'].mean()
mediana = df_clientes['idade'].median()

ax.axvline(media, color='red', linestyle='dashed', linewidth=1, label='Média')
ax.axvline(mediana, color='yellow', linestyle='dashed', linewidth=1, label='Mediana')

# Adicionando legenda
ax.legend()

# Ajustando layout para evitar sobreposição
plt.tight_layout()

# Mostrando os gráficos
plt.show()
```

```python
# Criar uma cópia explícita do DataFrame
df_clientes_original = df_clientes.copy()

# Definir as regras e valores das classes econômicas de acordo com a FGV e multiplicando por 12 para ter a renda anual
regras_classe_economica = [
    (df_clientes['renda_anual'] <= 1254*12),       # Classe E
    (df_clientes['renda_anual'] <= 2004*12),       # Classe D
    (df_clientes['renda_anual'] <= 8640*12),       # Classe C
    (df_clientes['renda_anual'] <= 11261*12),      # Classe B
    (df_clientes['renda_anual'] > 11261*12)        # Classe A
]

valores_classe_economica = ['E', 'D', 'C', 'B', 'A']

# Aplicar as regras e criar a nova coluna 'classe_economica'
df_clientes['classe_economica'] = pd.Categorical(np.select(regras_classe_economica, valores_classe_economica), 
                                                            categories=valores_classe_economica, ordered=True)

# Mostrar o DataFrame com a nova coluna
df_clientes
```

#### Vamos verificar se realmente a clientela da Gelateria Lillo é Classe B+ (Considerando Classe A e B).

```python
# Contagem de cada classe econômica
contagem_classes = df_clientes['classe_economica'].value_counts().sort_index(ascending=False)

# Exibir a contagem de cada classe econômica
print(contagem_classes)
```

#### De acordo com a FGV [Classes Econômicas](https://cps.fgv.br/qual-faixa-de-renda-familiar-das-classes), a Gelateria Lillo possui seu clientes classificados majoritariamento dentro da Classe C e os demais na Classe B, não sendo observados clientes na classe A, D e E. 

#### Vamos verificar o comportamento do cadastro dos clientes em relação aos meses e anos para verificar o comportamento e se há alguma tendência em relação ao clima de acordo com as estações.

```python
# Converter a coluna 'membro_desde' para o tipo datetime
df_clientes['membro_desde'] = pd.to_datetime(df_clientes['membro_desde'], format='%Y%m%d')

# Extrair os dados de ano, mês e dia de cadastro
df_clientes['ano_cadastro'] = df_clientes['membro_desde'].dt.year
df_clientes['mes_cadastro'] = df_clientes['membro_desde'].dt.month
df_clientes['dia_cadastro'] = df_clientes['membro_desde'].dt.day

# Exibir o DataFrame resultante
df_clientes
```

```python
# Contagem de clientes cadastrados por ano
contagem_por_ano = df_clientes['ano_cadastro'].value_counts().sort_index()

# Criando o gráfico de barras para verificar quanto clientes se cadastraram por ano
plt.figure(figsize=(6, 4))
contagem_por_ano.plot(kind='bar', rot=0)
plt.title('Número de Clientes Cadastrados por Ano')
plt.xlabel('Ano de Cadastro')
plt.ylabel('Número de Clientes')
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

**E como esse cadastrado deve ter ocorrido ao longo dos meses para cada ano?**

```python
# Contagem de clientes cadastrados em cada ano por mês
contagem_por_ano_mes = df_clientes.groupby(['ano_cadastro', 'mes_cadastro']).size().unstack(fill_value=0)

# Criar o gráfico de barras
contagem_por_ano_mes.plot(kind='bar', figsize=(12, 6))
plt.title('Número de Clientes Cadastrados em Cada Ano por Mês')
plt.xlabel('Ano de Cadastro')
plt.ylabel('Número de Clientes Cadastrados')
plt.legend(title='Mês', labels=[f'Mês {i}' for i in range(1, 13)], loc='upper left')
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

### Entendendo a Problemática

Precisamos encontrar uma maneira de determinar a eficácia de diferentes ofertas para cada cliente. Essa análise é viável devido à ordem cronológica dos eventos registrados. Quando consideramos uma oferta como eficaz (ou seja, o cliente a usou para realizar uma compra), os eventos que levam o cliente a converter a oferta em uma compra podem ocorrer de duas formas:

1. Recebimento da oferta -> Visualização da oferta -> Transação -> Conclusão da oferta (aqui tratamos ofertas de cupom e desconto);

2. Recebimento da oferta -> Visualização da oferta -> Transação (aqui tratamos ofertas informativas, onde as transações devem ocorrer durante o período da oferta)

Apenas quando essas duas sequências de eventos de conversão são observadas, consideramos a oferta como eficaz.

Além disso, existem cenários que caracterizam ofertas ineficazes:

- Oferta recebida, mas nenhum evento subsequente ocorre (a oferta não é visualizada nem utilizada);
- Recebimento da oferta -> Transação -> Visualização da oferta (nesse caso, o cliente não usou a oferta na transação, comprando antes de vê-la);
- Transação sem oferta (clientes desse tipo realizam transações independentemente de ofertas, tornando menos vantajoso enviar ofertas a eles do ponto de vista financeiro da empresa).

### Clusterização e Segmentação dos Clientes de acordo com os Eventos relacionados e Ofertas definidas

```python
# pra começar, vamos preencher com 0 os NAs da base com a qual iremos trabalhar
df_eventos_ofertas.fillna(0, inplace=True)

# Juntar as informações relevantes dos DataFrames
df_clustering = df_clientes.merge(df_eventos_ofertas, on='id_cliente', how='inner')
```

```python
df_clustering.head(10)
```

```python
df_clustering.info()
```

```python
# Selecionar colunas categóricas para aplicar o OneHotEncoder
categorical_columns = ['genero', 'classe_economica', 'tipo_evento', 'oferta']

# Converter as colunas para o tipo str
for col in categorical_columns:
    df_clustering[col] = df_clustering[col].astype(str)

# Criar o objeto OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Aplicar o OneHotEncoder nas colunas selecionadas
encoded_features = encoder.fit_transform(df_clustering[categorical_columns])

# Obter os nomes das colunas codificadas
encoded_columns = encoder.get_feature_names_out(categorical_columns)

# Criar um DataFrame com as features codificadas
encoded_df = pd.DataFrame(encoded_features, columns=encoded_columns)

# Concatenar o DataFrame codificado com o DataFrame original
df_clustering_encoded = pd.concat([df_clustering, encoded_df], axis=1)

# Exibir as primeiras linhas do DataFrame com as features codificadas
df_clustering_encoded.head()
```

```python
df_clustering_encoded.info()
```

```python
# Selecionar apenas colunas numéricas do DataFrame
numeric_columns = df_clustering_encoded.select_dtypes(include=[np.number])

# Aplicar StandardScaler para padronização
scaler = StandardScaler()
data_scaled = scaler.fit_transform(numeric_columns)

# Definir um range de valores para o número de clusters
min_clusters = 2
max_clusters = 10

# Lista para armazenar as inércias
inertia_values = []

# Loop para variar o número de clusters
for n_clusters in range(min_clusters, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    kmeans.fit(data_scaled)
    inertia_values.append(kmeans.inertia_)

# Plotar o gráfico da inércia em relação ao número de clusters
plt.figure(figsize=(10, 6))
plt.plot(range(min_clusters, max_clusters + 1), inertia_values, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Inércia')
plt.title('Elbow Method para Determinar o Número Ideal de Clusters')
plt.show()
```

```python
def calculate_wcss(df_clustering_encoded, max_clusters=10):
    wcss = []
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(df_clustering_encoded)
        wcss.append(kmeans.inertia_)
    return wcss

def optimal_number_of_clusters(wcss):
    diff1 = np.diff(wcss, 1)
    diff2 = np.diff(wcss, 2)
    n_clusters = np.argmax(diff2) + 2
    return n_clusters

# Selecionar as colunas numéricas relevantes para a clusterização
numeric_columns = ['idade', 'renda_anual', 'valor', 'recompensa_recebida', 'recompensa_definida',
                   'valor_minimo', 'duracao', 'web', 'email', 'mobile', 'social', 'oferta_compre 1, leve 2',
                   'oferta_desconto', 'oferta_informativo']

# Padronizar as colunas numéricas
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_clustering_encoded[numeric_columns])

# Calcular as somas dos quadrados para diferentes números de clusters
wcss_values = calculate_wcss(data_scaled, max_clusters=10)

# Encontrar o número ótimo de clusters usando o método Elbow
optimal_clusters = optimal_number_of_clusters(wcss_values)
print("Número ótimo de clusters:", optimal_clusters)

# Plotar o gráfico de Elbow
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss_values, marker='o')
plt.title('Método Elbow para Determinar Número Ótimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inércia')
plt.xticks(range(1, 11))
plt.show()
```

```python
# Selecionar apenas colunas numéricas do DataFrame
df_clustering_encoded = df_clustering_encoded.select_dtypes(include=[np.number])

# Realizar a clusterização com o número ótimo de clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
clusters = kmeans.fit_predict(df_clustering_encoded)

# Adicionar os rótulos dos clusters ao DataFrame
df_clustering_encoded['cluster'] = clusters

# Visualização dos clusters para número de clusters igual a 2
plt.figure(figsize=(10, 8))
plt.scatter(df_clustering_encoded.iloc[:, 0], df_clustering_encoded.iloc[:, 1], c=clusters, cmap='rainbow', marker='o', edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', marker='X', label='Centroids')
plt.title('Visualização dos Clusters (Número de Clusters = 2)')
plt.xlabel('Idade')
plt.ylabel('Renda Anual')
plt.legend()
plt.show()
```
